# Importing libraries
let's use pandas and numpy

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Some styling choices for graphs

```python
sns.set_style('darkgrid')
sns.set_palette('deep')
```

# Reading the dataset file

```python
cereal_df = pd.read_csv("Cereals.csv")
cereal_df
```

# Check for NaN values & getting information for each columns

```python
print('NaN values:', cereal_df.isnull().values.any())
cereal_df.info()
```

# Statistics description of the columns

```python
cereal_df.describe()
```

# Convert any -1 value to NaN

```python
# Convert -1 values to NaN
cereal_df = cereal_df.applymap(lambda x: np.nan if x == -1 else x)
# use panda's .drpna()with the default values to drop any row with NaN values
cereal_df = cereal_df.dropna(axis='index', how= 'any')
cereal_df
```

```python
# Now let's focusing on dropping shelf and rating since they are irrelevant to our study

cereal_df = cereal_df.drop(['shelf','rating'], axis=1)
```

From the results above we can conclude that the cereals are defined as one serving and have for weight of 1.0 ounce. Most of them having different serving size by cups or weight, we have to come up with a serving size; let's normalize the sample information to 1.0 ounce of weight.

```python
print('weight \n', cereal_df['weight'].value_counts(), '\n')
print('cups\n', cereal_df['cups'].value_counts())
```

```python
# Let's copy the dataset for further consideration
non_normal_df = cereal_df.copy()
# lets create a new dataset from only numerical columns
num_df = cereal_df.select_dtypes(include=[np.number])
# let's normalize each row by its corresponding weight to 1.0 ounce
weight_norm_df = num_df.div(num_df['weight'], axis ='rows')
# Lets check our normalization 
print('weight \n', weight_norm_df['weight'].value_counts())
```

```python
# let's replace the original numerical columns with the weight normalized values
cereal_df[weight_norm_df.columns]= weight_norm_df
# lets drop the weight and cups columns. All weight values must be 1.0 and cups are irrelevant
cereal_df.drop(columns=['weight','cups'], axis=1, inplace=True)
```

```python
# let's have a graph
sns.set_context('talk', font_scale = 1.6) 
cereal_g = sns.PairGrid(cereal_df, corner=True)
# Different type of plot to use on the diagonal
cereal_g.map_diag(plt.hist)
# type of plot for lower half of the frid
cereal_g.map_lower(plt.scatter)
```

```python
# let's round up the values to be readable
cereal_df['vitamins'] = cereal_df['vitamins'].round(1)
```

```python
# Lets size seaborn
sns.set_context('paper', font_scale = 1.8)
```

```python
# Let's call Pairgrid
cereal_g = sns.PairGrid(cereal_df, hue = 'vitamins',x_vars=['fiber','protein'],y_vars=['potass'],height=6, aspect=1, palette='hls',)
# Let's use scatter plots for our grid
cereal_g.map(plt.scatter,**{'edgecolor': 'k', 's':80, 'linewidth': 1.6})
cereal_g.add_legend(title ='Vitamins (% of Daily)')
# Lets labels
cereal_g.axes[0,0].set_xlabel('Fiber (grams)')
cereal_g.axes[0,1].set_xlabel('Protein (grams)')
cereal_g.axes[0,0].set_ylabel('Potassium (miligrams)')
plt.savefig('Good_variables.png')


```
